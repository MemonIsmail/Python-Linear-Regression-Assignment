```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('Data.csv')
df.head()
df.info()
```

```python
plt.title('Time Spent in Hours on Social Media')
sns.distplot(df['time_spent'])
plt.show()
```

```python
plt.scatter(df['age'], df['time_spent'], color = 'lightcoral')
plt.title('Age VS Time Spent')
plt.xlabel('Age')
plt.ylabel('Time Spent in Hours')
plt.box(True)
plt.show()
```

```python
X = df[['age']]
y = df['time_spent']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred_test = regressor.predict(X_test)     
y_pred_train = regressor.predict(X_train)   
```

```python
# Prediction on training set
plt.scatter(X_train, y_train, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Age VS Time Spent (Training Set)')
plt.xlabel('Age')
plt.ylabel('Time Spent in Hours')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Age/Time_Spent', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
plt.scatter(X_test, y_test, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Age VS Time Spent (Test Set)')
plt.xlabel('Age')
plt.ylabel('Time Spent in Hours')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Age/Time_Spent', loc='best', facecolor='white')
plt.box(False)
plt.show()
```
